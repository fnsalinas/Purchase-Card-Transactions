### Loading all xls files in a dataframe list

```python
df_list = []
for file in xls:
    print(f'{dt.now().strftime("%H:%M:%S")}: Working on {file}')
    df_sheets = pd.read_excel(file, sheet_name=None)
    for sheet in df_sheets.keys():
        df_sheet = df_sheets[sheet]
        df_sheet['file_name'] = file.split('\\')[-1]
        df_sheet['sheet_name'] = sheet
        if df_sheet.shape[0]>0: df_list.append(df_sheet)

print(f'{dt.now().strftime("%H:%M:%S")}: Done!')
print(f'Dataframes count: {len(df_list)}')
```

### Identifying all column names

There are dataframes with 13, 14, 16 and 18 columns, in the next step we will consolidate one table under the other by column names.

```python
df_columns = pd.DataFrame([[x.loc[0,'file_name'].split('\\')[-1][-30:]] + list(x.columns) for x in df_list])
df_columns['ncols'] = df_columns.apply(lambda x: len([y for y in x if y != None])-1, axis=1)
print(df_columns.groupby('ncols')[0].count())
df_columns.head()
```

```python
variables = []
for c in range(1, len(df_columns.columns)-1):
    for r in df_columns.index:
        variables.append(df_columns.iloc[r,c])

variables = list(dict.fromkeys(variables))
print(f'Number of variables: {len(variables)}')
variables
```

```python
vars_rename = {
    'file_name': 'file_name',
    'sheet_name': 'sheet_name',
    'TRANS DATE': 'trans_date',
    'TRANS VAT DESC': 'trans_vat_desc',
    'ORIGINAL GROSS AMT': 'original_gross_amt',
    'ORIGINAL CUR': 'original_cur',
    'MERCHANT NAME': 'merchant_name',
    'CARD NUMBER': 'card_number',
    'BILLING GROSS AMT': 'billing_gross_amt',
    'BILLING CUR CODE': 'billing_cur_code',
    'BILLING CUR CODE.1': 'billing_cur_code_1',
    'TRANS TAX AMT': 'trans_tax_amt',
    'TRANS CAC CODE 1': 'trans_cac_code_1',
    'TRANS CAC DESC 1': 'trans_cac_desc_1',
    'TRANS CAC CODE 2': 'trans_cac_code_2',
    'TRANS CAC DESC 2': 'trans_cac_desc_2',
    'TRANS CAC CODE 3': 'trans_cac_code_3',
    'Directorate ': 'directorate',
    'Directorate': 'directorate',
    'Directorates': 'directorate',
    'Unnamed: 10': 'unnamed'
}
```

**Note**: There are similar column names in diferent dataframes like "Directorate ", "Directorate" and "Directorates", we will take care on the consolidation step to rename the columns and consolidate in the same column.

```python
df_list = [x.rename(columns=vars_rename) for x in df_list]
```

### Consolidating datarames

```python
df_concat = pd.concat(df_list, axis=0, ignore_index=True)

cols = list(dict.fromkeys(list(vars_rename.values())))
df_concat = df_concat[cols].copy()

df_concat.info()
df_concat.head()
```

### Saving consolidated dataframe Consolidating

```python
df_concat.to_csv('../data/interim/raw_consolidated_data.csv', sep='~', encoding='utf-8', index=False)
```
